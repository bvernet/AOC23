```python
#parsing
lines = puzzle.input_data.split('\n')

parsed_lines = [line.split(": ")[1] for line in lines]
```

```python
lines
```

```python
result = 0
schema = r"(?:(\d+) (red|green|blue))(?:, (\d+) (red|green|blue))?(?:, (\d+) (red|green|blue))?"
for id, line in enumerate(parsed_lines):
    possibleGame = True
    match = re.findall(schema, line)
    for set in match:
        nb_red = 0
        nb_green = 0
        nb_blue = 0
        for i in [1,3,5]:
            match set[i]:
                case 'red':
                    nb_red = int(set[i-1])
                case 'green':
                    nb_green = int(set[i-1])
                case 'blue':
                    nb_blue = int(set[i-1])
        if not(nb_red < 13 and nb_green < 14 and nb_blue < 15):
            possibleGame = False
    if possibleGame:
        result += id +1
```
