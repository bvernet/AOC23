### Part 1

```python
import regex as re
puzzle = Puzzle(2023, 2)
```

```python
#parsing
lines = puzzle.input_data.split('\n')
parsed_lines = [line.split(": ")[1] for line in lines]
```

```python
result = 0
schema = r"(?:(\d+) (red|green|blue))(?:, (\d+) (red|green|blue))?(?:, (\d+) (red|green|blue))?"
for id, line in enumerate(parsed_lines):
    possibleGame = True
    match = re.findall(schema, line)
    for set in match:
        nb_red = 0
        nb_green = 0
        nb_blue = 0
        for i in [1,3,5]:
            match set[i]:
                case 'red':
                    nb_red = int(set[i-1])
                case 'green':
                    nb_green = int(set[i-1])
                case 'blue':
                    nb_blue = int(set[i-1])
        if not(nb_red < 13 and nb_green < 14 and nb_blue < 15):
            possibleGame = False
    if possibleGame:
        result += id +1
```

```python
print(result) #1867
```

```python
puzzle.answer_a = result #1867
```

### Part 2

```python
result = 0
schema = r"(?:(\d+) (red|green|blue))(?:, (\d+) (red|green|blue))?(?:, (\d+) (red|green|blue))?"

for id, line in enumerate(parsed_lines):
    nb_red = 0
    nb_green = 0
    nb_blue = 0
    match = re.findall(schema, line)
    for set in match:
        nb_red_set = 0
        nb_green_set = 0
        nb_blue_set = 0
        for i in [1,3,5]:
            match set[i]:
                case 'red':
                    nb_red_set = int(set[i-1])
                case 'green':
                    nb_green_set = int(set[i-1])
                case 'blue':
                    nb_blue_set = int(set[i-1])
        if nb_red_set > nb_red :
            nb_red = nb_red_set
        if nb_green_set > nb_green :
            nb_green = nb_green_set
        if nb_blue_set > nb_blue:
            nb_blue = nb_blue_set
    power_set = nb_red * nb_green * nb_blue
    result += power_set
```

```python
print(result)
```

```python
puzzle.answer_b = result #84538
```
